```python
my_dict = {
    "name" : ["Fuzzy", "Wuzzy"],
    "age": 9,
    "animal" : "dog",
    "fur color" : "brown"
}

my_dict["eye color"] = "brown"

print(my_dict)
print(f"{my_dict["name"]}")

try:
  print(my_dict["hair"])
except KeyError as ke:
  print("key not found", ke)
```

```python
# loop on keys
for k in my_dict:
  print(f'key: {k} | value: {my_dict[k]}')

# loop on values
for v in my_dict.values():
  print(f'our value is {v}')

for k,v in my_dict.items():
  print(f'keys : {k} values : {v}')
```

```python
# get .method()
# given a key return a value
# set a default if no key is present

my_dict.get("last name", "No value there")
```

```python
# .pop() method
my_dict["eye color"] = "brown"
my_dict.pop("eye color")
print(my_dict)
print(my_dict.pop("nails", "no value present"))
```

```python
my_dict2 = {
    "school" : "LAS",
    1601 : "address",
    "num_students" : 100
}

my_dict3 = my_dict | my_dict2
print(my_dict3)

my_dict4 = { "course" : "MCON232",
            "professor" : "Spiegel-Cohen"}

my_dict3 |= my_dict4
print(my_dict3)
```

```python
# deep copies
import copy
my_dict_copy = copy.deepcopy(my_dict3)
print(my_dict_copy)
```

```python
#comprehensions
dict_new = {key : value for key, value in my_dict.items()}
print(dict_new)

dict_nums = {
    1: "one",
    2: "two",
    3: "three",
    4: "four"
}

dict_even = {key : value for key, value in dict_nums.items() if key % 2 == 0}
print(dict_even)
```
